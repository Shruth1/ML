```python
#A1 encoding scheme
import pandas as pd

excel_file = "thyroid.xlsx"
data = pd.read_excel(excel_file)
#"oridinal" for ordinal variables and "nominal" for nominal variables
encoding_section = {
    'sex': 'ordinal',
    'on thyroxine': 'nominal',
    'query on thyroxine': 'nominal',
    'sick': 'nominal',
    'pregnant': 'nominal',
    'thyroid surgery': 'nominal',
    'I131 treatment': 'nominal',
    'query hypothyroid': 'nominal',
    'lithium': 'nominal',
    'goitre': 'nominal',
    'tumor': 'nominal',
    'hypopituitary': 'nominal',
    'psych': 'nominal',
    'TSH measured': 'nominal',
    'T3 measured': 'nominal',
    'TT4 measured': 'nominal',
    'T4U measured': 'nominal',
    'FTI measured': 'nominal',
    'TBG measured': 'nominal',
    'referral source': 'nominal',
    'Condition': 'nominal'
}

encoded_df = pd.DataFrame()

for column, section in encoding_section.items():
    if section == 'ordinal':
        encoded_df[column] = pd.factorize(data[column])[0]
    elif section == 'nominal':
        encoded_df = pd.concat([encoded_df, pd.get_dummies(data[column], prefix=column)], axis=1)#get_dummies for data manipulation
    else:
        pass

non_categorical_columns = ['Record ID', 'age', 'TSH', 'T3', 'TT4', 'T4U', 'FTI', 'TBG']
encoded_df = pd.concat([encoded_df, data[non_categorical_columns]], axis=1)
print(encoded_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

excel_file = "thyroid.xlsx"
data = pd.read_excel(excel_file)

numerical_columns = ['age', 'TSH', 'T3', 'TT4', 'T4U', 'FTI']

#converting non numericals to nan
for column in numerical_columns:
    data[column] = pd.to_numeric(data[column], errors='coerce')

#box plot
plt.figure(figsize=(12, 8))
data[numerical_columns].boxplot(sym='r+', vert=False)
plt.title('Box Plot of Numerical Columns')
plt.xlabel('Value')
plt.show()

#using z scores
outliers = {}
for column in numerical_columns:
    z_scores = stats.zscore(data[column])
    threshold = 3
    potential_outliers = (z_scores > threshold) | (z_scores < -threshold)
    outliers[column] = data[potential_outliers]

for column, outlier_data in outliers.items():
    print(f"Potential outliers in {column}:")
    print(outlier_data)

#IQR
Q1 = data[numerical_columns].quantile(0.25)
Q3 = data[numerical_columns].quantile(0.75)
IQR = Q3 - Q1

#lower and upper bounds for potential outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

iqr_outliers = {}
for column in numerical_columns:
    potential_outliers = (data[column] < lower_bound[column]) | (data[column] > upper_bound[column])
    iqr_outliers[column] = data[potential_outliers]

for column, outlier_data in iqr_outliers.items():
    print(f"Potential outliers in {column} (IQR method):")
    print(outlier_data)
```

```python
#A4 Similarity measure
import pandas as pd

excel_file = "thyroid.xlsx"
data = pd.read_excel(excel_file)

binary_attributes = ['on thyroxine', 'query on thyroxine', 'sick', 'pregnant', 'thyroid surgery',
                     'I131 treatment', 'query hypothyroid', 'query hyperthyroid', 'lithium', 'goitre',
                     'tumor', 'hypopituitary', 'psych', 'TSH measured', 'T3 measured', 'TT4 measured',
                     'T4U measured', 'FTI measured', 'TBG measured']

vector1 = data.loc[0, binary_attributes].values
vector2 = data.loc[1, binary_attributes].values

f11 = sum((vector1 == 1) & (vector2 == 1))
f01 = sum((vector1 == 0) & (vector2 == 1))
f10 = sum((vector1 == 1) & (vector2 == 0))
f00 = sum((vector1 == 0) & (vector2 == 0))

JC = f11 / (f01 + f10 + f11)

SMC = (f11 + f00) / (f00 + f01 + f10 + f11)

print(f"Jaccard Coefficient (JC): {JC}")
print(f"Simple Matching Coefficient (SMC): {SMC}")

if JC > SMC:
    print("Jaccard Coefficient is more appropriate for this comparison.")
elif SMC > JC:
    print("Simple Matching Coefficient is more appropriate for this comparison.")
else:
    print("Both coefficients are equally appropriate for this comparison.")
```

```python
#A5 cosine similarity measure
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

excel_file = "thyroid.xlsx"
data = pd.read_excel(excel_file)

data = data.replace('?', np.nan)

numerical_columns = ['age', 'TSH', 'T3', 'TT4', 'T4U', 'FTI']

data = data.dropna(subset=numerical_columns)

data[numerical_columns] = data[numerical_columns].astype(float)

observation_1 = data.iloc[0][numerical_columns].values.reshape(1, -1)
observation_2 = data.iloc[1][numerical_columns].values.reshape(1, -1)

cosine_sim = cosine_similarity(observation_1, observation_2)

print(f"Cosine Similarity between the two observations: {cosine_sim[0][0]}")
```

```python
#A6 Heatmap plot
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import jaccard_score as js
from sklearn.metrics.pairwise import cosine_similarity as cns

vectors = np.random.randint(2, size=(20, 10))
num_vectors = len(vectors)
jc_matrix = np.zeros((num_vectors, num_vectors))
smc_matrix = np.zeros((num_vectors, num_vectors))
cos_matrix = np.zeros((num_vectors, num_vectors))

for i in range(num_vectors):
    for j in range(num_vectors):
        if i != j:
            # Jaccard coefficient
            jc_matrix[i, j] = js(vectors[i], vectors[j])

            # Simple Matching Coefficient
            smc_matrix[i, j] = np.sum(np.logical_and(vectors[i], vectors[j])) / np.sum(np.logical_or(vectors[i], vectors[j]))

            # Cosine similarity
            cos_matrix[i, j] = cns([vectors[i]], [vectors[j]])[0, 0]

labels = [f"Vector {i+1}" for i in range(num_vectors)]

fig, axes = plt.subplots(3, 1, figsize=(10, 15))

sns.heatmap(jc_matrix, annot=True, xticklabels=labels, yticklabels=labels, ax=axes[0])
axes[0].set_title('Jaccard Coefficient')

sns.heatmap(smc_matrix, annot=True, xticklabels=labels, yticklabels=labels, ax=axes[1])
axes[1].set_title('Simple Matching Coefficient')

sns.heatmap(cos_matrix, annot=True, xticklabels=labels, yticklabels=labels, ax=axes[2])
axes[2].set_title('Cosine Similarity')

plt.tight_layout()
plt.show()
```
